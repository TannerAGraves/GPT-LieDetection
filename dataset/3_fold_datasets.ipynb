```python
import numpy as np
import pandas as pd
```

```python
dc = pd.read_csv('everything.csv', sep=',', encoding='UTF-8')#encoding='ISO-8859-1')
```

```python
dc.groupby('set').count()
```

```python
#separate the 3 datasets
A=dc[dc['set']=='A']
B=dc[dc['set']=='B']
C=dc[dc['set']=='C']
#print(A.iloc[0:100])
```

```python
seed=123
#shuffle
A_shuf=A.sample(frac=1,random_state=seed,replace=False).reset_index()
#take different slices of the shuffle
A1=A_shuf.iloc[0:100]
A2=A_shuf.iloc[100:200]
A3=A_shuf.iloc[200:300]
B_shuf=B.sample(frac=1,random_state=seed,replace=False).reset_index()
B1=B_shuf.iloc[0:100]
B2=B_shuf.iloc[100:200]
B3=B_shuf.iloc[200:300]
C_shuf=C.sample(frac=1,random_state=seed,replace=False).reset_index()
C1=C_shuf.iloc[0:100]
C2=C_shuf.iloc[100:200]
C3=C_shuf.iloc[200:300]
#Folds
F1=pd.concat([A1,B1,C1])
F2=pd.concat([A2,B2,C2])
F3=pd.concat([A3,B2,C3])
#Training sets
Tr1=pd.concat([F1,F2])
Tr2=pd.concat([F1,F3])
Tr3=pd.concat([F2,F3])
#Validation sets
Val1=pd.concat([F3])
Val2=pd.concat([F2])
Val3=pd.concat([F1])
#Test set (remaining samples)
Test=pd.concat([A_shuf.iloc[300:],B_shuf.iloc[300:],C_shuf.iloc[300:]])
```

```python
CV_test=pd.concat([A_shuf.iloc[300:500],B_shuf.iloc[300:500],C_shuf.iloc[300:500]])
CV_test.to_json('CV_test_final.json', orient='records', lines=True)
```

```python
renameCols = {'text':'prompt','label':'target'}
F1_out = F1[['text','label','set']].rename(columns=renameCols)
F1_out['fold'] = 1
F1_out['target'] = F1_out['target'] == 1
F2_out = F2[['text','label','set']].rename(columns=renameCols)
F2_out['fold'] = 2
F2_out['target'] = F2_out['target'] == 1
F3_out = F3[['text','label','set']].rename(columns=renameCols)
F3_out['fold'] = 3
F3_out['target'] = F3_out['target'] == 1
folds_out = pd.concat([F1_out,F2_out,F3_out])
folds_out.to_csv('FOLDS.csv')
folds_out
```
